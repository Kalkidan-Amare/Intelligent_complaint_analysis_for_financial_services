This notebook runs representative questions through the RAG pipeline and inspects retrieved sources.

Prerequisite: point `FULL_CHROMA_DIR` to the provided full-scale persisted ChromaDB directory.

```python
from src.rag.pipeline import RAGPipeline

rag = RAGPipeline(collection_name='complaints_full')  # set FULL_CHROMA_DIR env var first

questions = [
  'Why are people unhappy with Credit Cards?',
  'What are common issues in Money Transfers?',
  'Are there fraud-related complaints in Savings Accounts?',
]

for q in questions:
    res = rag.ask(q, top_k=5)
    print('Q:', q)
    print('A:', res.answer)
    print('--- Sources (top 2) ---')
    for s in res.sources[:2]:
        meta = s.metadata or {}
        print(meta.get('product_category'), meta.get('complaint_id'), meta.get('issue'))
        print(s.text[:300].replace('\n',' '), '...')
    print('-' * 60)
```
